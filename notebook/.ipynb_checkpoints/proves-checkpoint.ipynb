```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

from collections import namedtuple #import the object namedtupla in order to save the data and optimize later actions 
```

```python
PosicionTiempo = namedtuple('PosicionTiempo', ('posicion', 'tiempo')) #we create the "shape" of the tuplas were we are saving the data
```

```python
def vert_shot(x_0, v_0, dt, g = -9.81):
    '''
    key args:
    x_0: initial position of the object
    v_0: initial velocity of the object
    dt: time step for the numerical calculations
    
    Returns: 
    A tuple with two lists of values: 'pos', which contains all the positions of the objects
    and 'temps', which contains the time associated with the positions.
    This values are calculated using a backward in time numerical calculation
    '''
    x = x_0
    v = v_0
    i = t = 0
    pos = []
    temps = []
    while x >= 0:
        i = i+1
        pos.append(x)
        temps.append(t)
        x = x + v * dt
        v = v + g * dt
        t = i * dt
    pos.append(x)
    temps.append(t)   
    
    return PosicionTiempo(pos, temps)
```

```python
# Tiro vertical forward in time

def vert_shot_ft(x_0, v_0, dt, g = -9.81):
    '''
    keyword args:
    x_0: initial position of the object
    v_0: initial velocity of the object
    dt: time step for the numerical calculations
    
    Returns: 
    A tuple with two lists of values: 'pos', which contains all the positions of the objects
    and 'temps', which contains the time associated with the positions.
    This values are calculated using a Forward in time numerical calculation, because we update the 'v' value
    before the position, thus we use v(n+1)
    '''
    x = x_0
    v = v_0
    i = t = 0
    pos = []
    temps = []
    while x >= 0:
        i = i+1
        pos.append(x)
        temps.append(t)
        v = v + g * dt
        x = x + v * dt
        t = i * dt
    pos.append(x)
    temps.append(t)   
    
    return PosicionTiempo(pos, temps)
```

```python
# Tiro vertical centered in time
def vert_shot_ct(x_0, v_0, dt, g = -9.81):
    '''
    keyword args:
    x_0: initial position of the object
    v_0: initial velocity of the object
    dt: time step for the numerical calculations
    
    Returns: 
    A tuple with two lists of values: 'pos', which contains all the positions of the objects
    and 'temps', which contains the time associated with the positions.
    This values are calculated using a Centered in time numerical calculation. 
    '''
    x = x_0
    v = v_0
    i = t = 0
    pos = []
    temps = []
    while x >= 0:
        i = i+1
        pos.append(x)
        temps.append(t)
        x = x + (v + 0.5 * g * dt) * dt
        v = v + g * dt
        t = i * dt
    pos.append(x)
    temps.append(t) 
    return PosicionTiempo(pos, temps)
```

```python
def vert_shot_analytic(x_0, v_0, dt, g = -9.81):
    '''
    keyword args:
    x_0: initial position of the object
    v_0: initial velocity of the object
    dt: time step for the numerical calculations
    
    Returns: 
    A tuple with two lists of values: 'pos', which contains all the positions of the objects
    and 'temps', which contains the time associated with the positions.
    This values are calculated using the analytical formula of the vertical shot
    '''
    x = x_0
    i = t = 0
    pos = []
    temps = []
    pos.append(x)
    temps.append(t)
    while x >= 0:
        i = i+1
        t = i * dt
        x = v_0 * t + 0.5 * g * t**2
        pos.append(x)
        temps.append(t)
    
    return PosicionTiempo(pos, temps)
```

```python
'''
Here we define some values to simplify the treatment and execution of the function
'''
x_0 = 0
v_0 = 25
dt = 0.1
dt_2 = 0.5
dt_3 = 0.01
```

```python
'''
Now we execute every function using the same initial values for three different dt,
but the analytical solution, wich always uses dt = 0.01
'''
```

```python
pt1 = vert_shot(x_0, v_0, dt)
```

```python
pt2 = vert_shot_ft(x_0, v_0, dt)
```

```python
pt3 = vert_shot_ct(x_0, v_0, dt)
```

```python
pta = vert_shot_analytic(x_0, v_0, dt_3)
```

```python
pt1_2 = vert_shot(x_0, v_0, dt_2)
```

```python
pt2_2 = vert_shot_ft(x_0, v_0, dt_2)
```

```python
pt3_2 = vert_shot_ct(x_0, v_0, dt_2)
```

```python
pt1_3 = vert_shot(x_0, v_0, dt_3)
```

```python
pt2_3 = vert_shot_ft(x_0, v_0, dt_3)
```

```python
pt3_3 = vert_shot_ct(x_0, v_0, dt_3)
```

```python
'''
Finally, we plot the results of the calculations in three different graphs using matplotlib tools
and setting labels for an appropiate interpretation
'''
```

```python
plt.title("Vertical Shot, dt = 0.1")
plt.xlabel("Time")
plt.ylabel("Position")
plt.plot(pt1.tiempo, pt1.posicion, label="Numerical Calc. BT")
plt.plot(pt2.tiempo, pt2.posicion, label="Numerical Calc. FT")
plt.plot(pt3.tiempo, pt3.posicion, label="Numerical Calc. CT")
plt.plot(pta.tiempo, pta.posicion, label = "Analytical Solution")
plt.legend()
```
